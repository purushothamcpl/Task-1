```python
df=df.dropna()
df
```

```python
df.drop_duplicates()
# If any duplicate rows present in the dataset will be removed.
```

```python
# date_added is a column in the dataset.
# Parse into datetime objects.
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Formating into dd-mm-yyyy.
df['date_standardized'] = df['date_added'].dt.strftime("%d-%m-%Y")

print(df[['date_added', 'date_standardized']])
```

```python
# Cleaning column headers: lowercase, strip spaces, replace spaces with underscores
df.columns = (
    df.columns
    .str.strip()            
    .str.lower()            
    .str.replace(' ', '_')  
    .str.replace('[^0-9a-zA-Z_]', '', regex=True)  
)

print(df.columns)

# removing leading/trailing space
# converting to lowercase
# replacing spaces with underscores
# removing special chars
```
